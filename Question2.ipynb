```python
#QUESTION-2
import pandas as pd,matplotlib.pyplot as plt,numpy as np
df=pd.read_csv("startup_funding.csv")
result={}
df=df["InvestorsName"]
df.dropna(inplace=True)
for row in df:
    investor_list=row.strip().split(",")
    for i in range(len(investor_list)):
        #Handling case-sensitiveness of Investor names.
        name=investor_list[i].strip().title()
        result[name]=result.get(name,0)+1
#Handling Undisclosed Investors.
del result["Undisclosed Investors"],result[""]
#Sorting Investors(Descending Order) on Number of Investments.
sorted_result=sorted(result.items(),key=lambda x:x[1],reverse=True)
#Top 5 Investors.
x_axis=list()
y_axis=list()
for element in sorted_result[:5]:
    x_axis.append(element[0])
    y_axis.append(element[1])
#Plotting Bar-Graph.
plt.bar(x_axis,y_axis)
plt.grid()
plt.title("TOP INVESTORS")
plt.xlabel("INVESTORS")
plt.ylabel("NUMBER OF FUNDINGS(January 2015-August 2017)")
plt.xticks(rotation=45)
plt.show()
#Actual Values
for i in range(len(x_axis)):
    print(x_axis[i],y_axis[i])
```
