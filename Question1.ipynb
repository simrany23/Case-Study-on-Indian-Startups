```python
#QUESTION-1
import pandas as pd,matplotlib.pyplot as plt
df=pd.read_csv("startup_funding.csv")
result={}
df["CityLocation"].dropna(inplace=True)
#For multiple city rows.
def separateCity(city):
    return city.split('/')[0].strip()
#First city is Indian.
df['CityLocation']=df['CityLocation'].apply(separateCity)
#Replacing "Delhi" with "New Delhi".
df['CityLocation'].replace("Delhi","New Delhi",inplace=True)
location_df=df["CityLocation"]
for city in location_df:
    #Taking care of case-sensitiveness of city names.
    city_name=str(city).title()
    if city_name=="Gurgaon" or city_name=="Noida" or city_name=="New Delhi":
        result["NCR"]=result.get("NCR",0)+1
    else:
        result[city_name]=result.get(city_name,0)+1
#Handling Nan city_name.
del result['Nan']
#Collecting data of relevant cities.
x_axis=list()
y_axis=list()
for city in result:
    if city=="NCR" or city=="Bangalore" or city=="Mumbai":
        x_axis.append(city)
        y_axis.append(result[city])
#Plotting Bar-Graph.
plt.bar(x_axis,y_axis)
plt.grid()
plt.title("INDIAN CITIES VS NUMBER OF FUNDINGS")
plt.xlabel("INDIAN CITIES")
plt.ylabel("NUMBER OF FUNDINGS(January 2015-August 2017)")
plt.show()
#Actual Values
for i in range(len(x_axis)):
    print(x_axis[i],y_axis[i])
```
